#1. Limpeza e Análise de Dados de **Vendas**

1.1 Importação de dataset de vendas

```python
import pandas as pd

file_path = '/content/dados_vendas.csv'
df_dados_vendas = pd.read_csv(file_path)
```

1.2 Visualização dos registros

```python
total_registros = len(df_dados_vendas)
print(f"\nTotal de registros no dataframe: {total_registros}\n")

pd.set_option('display.max_rows', None)
df_dados_vendas
```

1.3 Limpeza de dados

1.3.1 Remoção de registros com valores faltantes

```python
#verificaçao de valores faltantes
faltantes = df_dados_vendas[df_dados_vendas.isnull().any(axis = 1)]
print( "Registros com valores faltantes:\n", faltantes[["ID"]] )

#remoção dos registros com valores faltantes
df_faltantes = df_dados_vendas.dropna()

#total de registros
total_registros = len(df_faltantes)
print(f"\nTotal de registros pós remoçao de valores faltantes: {total_registros}\n")

#nova base de dados
pd.set_option('display.max_rows', None)
df_faltantes
```

1.3.2 Remoção de duplicatas

```python
# Verificacao de duplicatas
duplicatas = df_faltantes[df_faltantes.duplicated()]
print( "\nRegistros duplicados:\n", duplicatas[["ID"]] )

#quantidade de valores duplicados
duplicados = df_faltantes.duplicated().sum()
print(f"\nTotal de registros duplicados: {duplicados}")

#removendo de duplicatas
df_sem_duplicatas = df_faltantes.drop_duplicates()

#total de registros
total_registros = len(df_sem_duplicatas)
print(f"\nTotal de registros após remoção de duplicatas: {total_registros}\n")

#nova base de dados
pd.set_option('display.max_rows', None)
df_sem_duplicatas
```

1.3.3 Visualização geral pós limpeza

```python
#renomeando o dataframe
df_data_clean = df_sem_duplicatas

#mostrando informações
#quantidade de valores faltantes
total_faltantes = df_data_clean.isnull().sum().sum()
print(f"\nTotal de valores faltantes: {total_faltantes}")
print("\nTotal de duplicatas:", df_data_clean.duplicated().sum())

#total de registros
total_registros = len(df_data_clean)
print(f"\nTotal de registros no dataframe pós limpeza: {total_registros}\n")

#base de dados pós limpeza
pd.set_option('display.max_rows', None)
df_data_clean
```

1.4 Salvando base de dados limpa em CSV

```python
#exportando o DataFrame para um arquivo CSV
df_data_clean.to_csv("data_clean.csv", index=False)

#downlload do arquivo do Google Colab
from google.colab import files
files.download("data_clean.csv")
```

1.6 Cálculo do total de vendas (Quantidade * Preço)

```python
df_data_clean['Total_Vendas']= df_data_clean['Quantidade']* df_data_clean['Preco']

agrupado = df_data_clean.groupby('Produto').agg({
    'Quantidade':'sum','Preco': 'mean','Total_Vendas':'sum'}).reset_index( )

agrupado['Quantidade']= agrupado['Quantidade'].astype(int)

agrupado['Preco']= agrupado['Preco'].apply(
    lambda x: f"{x:,.2f}".replace(",","X").replace(".",",").replace("X","."))

agrupado['Total_Vendas']= agrupado['Total_Vendas'].apply (
    lambda x: f"{x:,.2f}".replace(",","X"). replace(".",",").replace("X","."))

print(agrupado[['Produto', 'Quantidade', 'Preco', 'Total_Vendas']] )
```

1.7 Produto com maior valor de total de vendas

```python
produto_maior_venda_total= agrupado.loc[agrupado['Total_Vendas'].str.replace(".","").str.replace(",", ".").astype(float).idxmax()]

print("Produto com o maior valor de total de vendas:")
print(produto_maior_venda_total)
```

1.8 Produto com mair número de vendas totais

```python
produto_mais_vendido = agrupado.groupby('Produto')['Quantidade'].sum().idxmax()
quantidade_mais_vendida = agrupado.groupby('Produto')['Quantidade'].sum().max()

print(f"Produto mais vendido: {produto_mais_vendido}")
print(f"Quantidade vendida do produto: {quantidade_mais_vendida}")
```

```python
# Salvando arquivo CSV
agrupado.to_csv("agrupado.csv", index=False, sep=',', encoding='utf-8')

# download do arquivo
from google.colab import files
files.download("agrupado.csv")
```
